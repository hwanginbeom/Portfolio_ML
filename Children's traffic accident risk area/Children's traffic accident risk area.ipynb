# 2.데이터 전처리

```python
# 9. 어린이보호구역
# CCTV설치여부는 N이나 CCTV설치대수가 1인 "세미초등학교" 결측치 처리 행 삭제

data[9].drop([x for x in (data[9][data[9]['시설명']=='세미초등학교'].index)])
data[9].reset_index(inplace=True, drop=True)
# data[9]
```

```python
# 9. 어린이보호구역(어린이집유치원) & 13. 어린이집유치원
# 어린이보호구역 데이터에 있는 어린이집유치원 리스트가 빠진 정보 추가해줘야한다
# 빠진 데이터를 data[35]로 저장

어린이보호구역_어린이집및유치원 = data[9][(data[9]['시설종류']=='어린이집') | (data[9]['시설종류']=='유치원')].reset_index(drop=True)

빠진_어린이집및유치원 = list(set(어린이보호구역_어린이집및유치원['시설명']) - set(data[13]['시설명']))
빠진_어린이집및유치원

index_list = []
for i in range(len(빠진_어린이집및유치원)) :
    for j in range(len(어린이보호구역_어린이집및유치원)) :
        if 빠진_어린이집및유치원[i] == 어린이보호구역_어린이집및유치원['시설명'][j] :
            index_list.append(j)

data[35] = 어린이보호구역_어린이집및유치원.iloc[index_list,:]
data[35].reset_index(inplace=True, drop=True)
# data[35]
```

```python
# 20. cctv
# CCTV 유형코드에서 " B"와 "BC" 결측치 처리 행 삭제

CCTV결측치인덱스 = data[20][(data[20]['CCTV 유형코드']=='BC') | (data[20]['CCTV 유형코드']==' B')].index
data[20].drop([x for x in (CCTV결측치인덱스)])
data[20].reset_index(inplace=True, drop=True)
# data[20]
```

```python
# 1. 주정차단속

# 결측치 제거
data[1].drop([x for x in (data[1][data[1]['단속위치_경도'] == '#N/A'].index)], inplace=True)
# 단속위치 중복제거
test = data[1].drop_duplicates(['단속위치']).reset_index(drop=True)
# test에 point 만들기
test['geometry'] = test.apply(lambda row : Point([ float(row['단속위치_경도']), float(row['단속위치_위도'])]), axis=1)

# 단속횟수 새로운 데이터프레임으로 만들기
df = pd.DataFrame(data=data[1]['단속위치'].value_counts().index, columns=['index'])
df['value'] = 0
for i in range(0, len(df)):
    df['value'][i] = data[1]['단속위치'].value_counts()[i]

test.reset_index(drop=True, inplace=True)
df.reset_index(drop=True, inplace=True)

test['단속횟수'] = 0
for x in range(0, len(df)):
    for y in range(0, len(test)):
        if df['index'][x] == test['단속위치'][y]:
            test['단속횟수'][y] = df['value'][x]

# 어린이교통사고에 데이터 넣기
어린이교통사고['주정차단속_개수'] = 0

for a in range(0, len(test)):
    for i in range(0, len(어린이교통사고)):
        if (test.geometry[a].within(어린이교통사고.geometry[i]) != False):
            어린이교통사고['주정차단속_개수'][i] += test['단속횟수'][a]
```
